# K-Means Clustering

- unsupervised learning - no labelled data
- K Means performs division of objects into clusters which are 'similar' between them and are 'dissimilar' to the objects belonging to another cluster
- k - number of clusters

***

## What is K-Means Clustering?
1. Initially, two centroids are assigned randomly.
2. The Euclidean distance is used to find out which centroid is closest to each data point and the data points are assigned to the corresponding centroids.
3. Reposition the two centroids for optimisation.
4. The process is iteratively repeated until our centroids become static (converged).

***

## Types of Clustering
Clustering:
- Hierarchical Clustering
    - Agglomerative
    - Divisive
- Partitional Clustering
    - K-Means
    - Fuzzy C-Means

***

**Hierarchical Clustering** - clusters have a tree-like structure or a parent-child relationship.

Agglomerative - 'Bottom-up' approach: Begin with each element as a separate cluster and merge them into successively large clusters.

Divisive - 'Top down' approach: Begin with the whole set and proceed to divide it into successively smaller clusters.

***

**Partitional Clustering** 

K-Means - Division of objects into clusters such that each object is in exactly one cluster, not several.

Fuzzy C-Means - Division of objects into clusters such that each object can belong to multiple clusters.

## Distance Measure:

1. Euclidean 
    - 'ordinary' straight line
    - distance between two points in Euclidean space

$$ d = \sqrt{\sum\limits_{i=1}^{n} {(q_i-p_i)^2}} $$

2. Squared Euclidean 
    - same equation as the Euclidean distance metric
    - but does not take the square root

$$ d = \sum\limits_{i=1}^{n} {(q_i-p_i)^2} $$

3. Manhanttan 
    - Simple sum of the horizontal and vertical components or the distance between two points measured along axes at right angles.

$$ d = \sum\limits_{i=1}^{n} {|q_x-p_x|+|q_y-p_y|} $$

4. Cosine 
    - Measures the angle between the two vectors

$$ d = \frac {\sum\limits_{i=0}^{n-1}(q_i-p_x)}{{\sum\limits_{i=0}^{n-1}(q_i)^2} \times {\sum\limits_{i=0}^{n-1}(p_i)^2}} $$

***

## How does K-Means clustering work?
1. Choose k (elbow point) - trial and error
    - How would you choose the optimum number of clusters?
    - Use Elbow method - within sum of squares (WSS) - sum of the squared distance between each member of the cluster and its centroid
    $$ WSS = \sum\limits_{i=1}^{m} (x_i-c_i)^2 $$ where $x_i$ = data point and $c_i$ = closest point to centroid
    - Now we draw a curve between WSS and the number of clusters.
    - When there is a slow change in the value of WSS, we should take that as elbow point.
2. Measure the distance
    - We can randomly initialise two points called c1 and c2 centroids.
    - Euclidean distance is a distance measure used to find out which data point is closest to our centroids.
3. Grouping based on minimum distance
    - Based upon the distance from c1 and c2 centroids, the data points will group itself into clusters.
4. Reposition the centroids
5. If clusters are unstable, back to (2.)
6. If clusters are stable, convergence!
    - Once the clusters become static, K-Means clustering algorithm is said to be converged. 

***

## K-Means Clustering Algorithm

<u>Step 1</u>: 
We randomly pick K cluster centers (centroids). Let's assume these are c1,c2,...,ck and we can say that $$ C = c_1,c_2,...,c_k $$ where C is the set of all centroids.

<u>Step 2</u>:
In this step, we assign each data point to closest center, this is done by calculating the Euclidean distance.

<u>Step 3</u>:
In this step, we find the new centroid by taking the average of all the points assigned to that cluster.

<u>Step 4</u>:
We repeat step 2 and 3 until none of the cluster assignment change.

## Demo 1

**Problem Statement**:
- Walmart wants to open a chain of stores across Florida and wants to find out the optimal store locations to maximise revenue.

**Solution**:
- Walmart already has a strong e-commerce presence.
- Walmart can use its online customer data to analyse the customer locations along with the monthly sales. 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set() # for plot styling
import numpy as np
```

```python
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
plt.scatter(X[:,0], X[:,1], s=50)
```

```python
from sklearn.cluster import KMeans 
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
y_kmeans
```

```python
from sklearn.metrics import pairwise_distances_argmin

def find_clusters(X, n_clusters, rseed=2):
    # 1. Randomly choose clusters
    rng = np.random.RandomState(rseed)
    i = rng.permutation(X.shape[0])[:n_clusters]
    centers = X[i]
    
    while True:
        # 2a. Assign labels based on closest center
        labels = pairwise_distances_argmin(X, centers)
        
        # 2b. Find new centers from mean of points
        new_centers = np.array([X[labels == i].mean(0) 
                                for i in range(n_clusters)])
        
        # 2c. Check for convergence 
        if np.all(abs(centers - new_centers)<3): 
            # original code is: centers == new_centers, however it might be an infinite loop
            break
        centers == new_centers
        
    return centers, labels

centers, labels = find_clusters(X, 4)
plt.scatter(X[:,0], X[:,1], c=y_kmeans, s=50, cmap='viridis')

plt.scatter(centers[:,0], centers[:,1], c='black', s=200, alpha=0.5)
```

## Demo 2 - Color Compression

**Problem Statement**:
- To perform color compression on images using K-Means algorithm

```python
from sklearn.datasets import load_sample_image
china = load_sample_image("flower.jpg")
ax = plt.axes(xticks=[],yticks=[])
ax.imshow(china);
```

```python
# Returns the dimension of the array
china.shape
```

**(427, 640, 3) means 427pixels x 640pixels x 3layers**

```python
# Pixel values
china
```

```python
# Reshape the data to [n_samples x n_featuers], and rescale the colors so that they lie between 0 and 1
data = china / 255.0 # use 0...1 scale
data = data.reshape(427 * 640, 3)
data.shape
```

**We changed a 3-dimensional to a 2-dimensional dataset.**

```python
data
```

**Now the values are between 0 and 1.**

```python
# Visualise these pixels in this colour space, using a subset of 10,000 pixel for efficiency.
def plot_pixels(data, title, colors=None, N=10000):
    if colors is None:
        colors = data
        
    # choose a random subset
    rng = np.random.RandomState(0)
    i = rng.permutation(data.shape[0])[:N]
    colors = colors[i]
    R, G, B = data[i].T
    
    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    ax[0].scatter(R, G, color=colors, marker='.')
    ax[0].set(xlabel='Red', ylabel='Green', xlim=(0,1), ylim=(0,1))
    
    ax[1].scatter(R, B, color=colors, marker='.')
    ax[1].set(xlabel='Red', ylabel='Blue', xlim=(0,1), ylim=(0,1))
    
    fig.suptitle(title, size=20);  
```

```python
plot_pixels(data, title='Input color space: 16 million possible colors')
```

```python
import warnings; warnings.simplefilter('ignore') # Fix NumPy issues.

# There are 16M colours, using the regular K Means would be very time-consuming, so we used MiniBatchKMeans
from sklearn.cluster import MiniBatchKMeans
kmeans = MiniBatchKMeans(16)
kmeans.fit(data)
new_colors = kmeans.cluster_centers_[kmeans.predict(data)]

plot_pixels(data, colors=new_colors, title="Reduced color space: 16 colors")
```

```python
china_recolored = new_colors.reshape(china.shape)

fig, ax = plt.subplots(1, 2, figsize=(16, 6), 
                      subplot_kw=dict(xticks=[], yticks=[]))
fig.subplots_adjust(wspace=0.05)
ax[0].imshow(china)
ax[0].set_title('Original Image', size=16)
ax[1].imshow(china_recolored)
ax[1].set_title('16-Color Image', size=16)
```
